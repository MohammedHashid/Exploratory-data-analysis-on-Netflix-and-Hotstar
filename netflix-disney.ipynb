<p style="font-size:170%;"> How much time do you spend to watch all the movies and TV shows that exist on each platform?

<p style = "font-size:117%;">There is a funny fact about this plataforms. We know that these platforms are intended to entertain. However, it is also known that we cannot stay all day watching. Despite being a funny fact, there is a little seriousness, because how much time in our lives will we spend watching movies and TV shows that will add almost nothing, or nothing, to our professional life, family life, personal life, social life, etc? 

<p style = "font-size:117%;"> So... how much time we will spend if we pretend watch all movies and tvshows? (obviously that it's an extreme case) 

<p style = "font-size:117%;"> First of all, i will set each movie and tvshows by the HOURS. We know that the most episodes on tv shows have a average of 45 min. Another thing that we need to now is the amount of ep have in one sesson... I'm going to guess that, the minimum, is 10 ep for the season per tvshows...
    
 <p style = "font-size:117%;"> Here i need to make a little ajust because some TV Shows and movies was NaN or Null. Because of this i drop some itens...

```python
# About Netflix

data_net_clean_NaN = data_netflix.drop(['show_id','cast','description','listed_in'], axis = 1).dropna()

series_net = data_net_clean_NaN[data_net_clean_NaN['duration'].str.contains('Season')]
filmes_net = data_net_clean_NaN[data_net_clean_NaN['duration'].str.contains('min')]

series_net_duration = []
filmes_net_duration = []

# Tv shows
for i in series_net['duration'].values:
    x = i.split()
    for ii in x[0]:
        xx = int(ii)
        math = ((xx*10)*45)/60
        series_net_duration.append(math)
        
# Movies
for i in filmes_net['duration'].values:
    x = i.split()
    for ii in x[:1]:
        xx = int(ii)/60
        filmes_net_duration.append(xx)

###########################################################################################################


# About Disney

data_disney_clean_NaN = data_disney.drop(['show_id','cast','description','listed_in'], axis = 1).dropna()

series_disney = data_disney_clean_NaN[data_disney_clean_NaN['duration'].str.contains('Season')]
filmes_disney = data_disney_clean_NaN[data_disney_clean_NaN['duration'].str.contains('min')]

series_disney_duration = []
filmes_disney_duration = []

# Tv shows
for i in series_disney['duration'].values:
    x = i.split()
    for ii in x[0]:
        xx = int(ii)
        math = ((xx*10)*45)/60
        series_disney_duration.append(math)
        
# Movies
for i in filmes_disney['duration'].values:
    x = i.split()
    for ii in x[:1]:
        xx = int(ii)/60
        filmes_disney_duration.append(xx)


###############################################################################################################

# Results of this

results_net = round(sum(series_net_duration)+sum(filmes_net_duration), 2)
results_disney = round(sum(series_disney_duration)+sum(filmes_disney_duration), 2)
```

```python
frame1 = pd.DataFrame(data = (results_disney, results_net), index=('Disney+', 'Netflix'))
frame1.plot.barh(figsize =(12,6))
plt.title('How much time you will spend ?', pad =15, fontsize = 18)
plt.legend(fontsize = 0)
plt.xlabel('Hours', fontsize=14)
plt.show()
```

<p style="font-size:125%;" > Results: It was to be expected that netflix had much more entertainment would take much longer to watch all of its programming.Different of Disney+. The end result was: Disney 1154.4 hours and Netflix with 11802.98 hours, u will speend more than 102.2% of time to watch Netflix compared with Disney+.
    
<p style="font-size:125%;"> Its important to remember that i needed make some ajust, i drop some values. Its means that you, certainly, will spend more hours watching the schedule that this graph can shows...
